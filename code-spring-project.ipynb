```python
# simple spring logic practice

import numpy as np
import matplotlib.pyplot as plt

#tool part--------------------------------------------------------------------------------------

def function(x): # for testing
    return np.sin(x)

def firstderi(f,x,dx) : # for first derivitation
    ans = (f(x)-f(x-dx))/dx
    return ans

def secondderi(f,x,dx): # for second derivitation
    ans = (firstderi(f,x+dx,dx)-firstderi(f,x,dx))/dx
    return ans

#-------------------------------------------------------------------------------------------------

def force_spring(k,x): # a spring
    return -k*x #x is the stretched length

#def spring(m1,m2,)

t = 0.1 #s
v = 0 #cm/s
start_pos = -5 # starting position cm
k = 2 # elastic constant
dt = 0.001 #s
mass = 0.5 #g
time_passed = 10

time_mark = np.arange(0.,time_passed,dt)
time_list = [] # time progress
force_list = [] # force function
accelaration_list = [] #accelaration function
velocity_list = [] # velocity function
position_list = [] #position function
xaxis_list = []

for t in time_mark : # for time 0~10s
    time_list.append(t)
    force_list.append(force_spring(k,start_pos))
    velocity_list.append(v)
    position_list.append(start_pos)
    a = force_spring(k,start_pos)/mass
    accelaration_list.append(a)
    v = v + a*dt
    start_pos = start_pos + v*dt
    t = t + dt
    xaxis_list.append(0)

plt.title("all stats of one spring")
plt.plot(time_list,force_list,label = 'force')
plt.plot(time_list,accelaration_list,label = 'acc')
plt.plot(time_list,velocity_list,label = 'velo')
plt.plot(time_list,position_list,label = 'pos')
plt.plot(time_list,xaxis_list,'black')
plt.legend()
plt.show()
```
